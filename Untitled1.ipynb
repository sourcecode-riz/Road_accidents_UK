```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from mlxtend.preprocessing import transactionencoder
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

con = sqlite3.connect("accident_data_v1.0.0_2023.db")
cur = con.cursor()
```

### Filtering accidents from 2020

```python
q = """
    SELECT a.*, v.*, c.*, l.*
    FROM accident a
    JOIN vehicle v
    ON a.accident_index = v.accident_index
    JOIN casualty c
    ON a.accident_index = c.accident_index
    JOIN lsoa l
    ON l.lsoa01cd = a.lsoa_of_accident_location
    WHERE a.accident_year = 2020;
"""
df = pd.read_sql(q, con)
df = df.loc[:, ~df.columns.duplicated()]
df.info()
```

### Data Cleaning

```python
# Initialize a dictionary to store columns with negative values
columns_with_negative_values = {}

# Iterate over each column
for column in df.columns:
    # Check if the column contains numeric values
    if np.issubdtype(df[column].dtype, np.number):
        # Count negative values in the column
        negative_count = (df[column] < 0).sum()
        # If there are negative values, add the column to the dictionary
        if negative_count > 0:
            columns_with_negative_values[column] = negative_count

# Print columns with negative values
if columns_with_negative_values:
    print("Columns with negative values:")
    for column, count in columns_with_negative_values.items():
        print(f"{column}: {count} negative values")
else:
    print("No columns contain negative values.")
```

### Replacing Negative Values of Important Columns

```python
#Replacing negative value of speed_limit
median_speed_limit = round(df['speed_limit'].median())
df.loc[df['speed_limit'] == -1, 'speed_limit'] = median_speed_limit
print("Median speed limit after replacement:", median_speed_limit)

#Replacing negative value of light_conditions
mean_light_condition = round(df['light_conditions'].mean())
df.loc[df['light_conditions'] == -1, 'light_conditions'] = mean_light_condition
print("Mean light condition after replacement:", mean_light_condition)

#Replacing negative value of weather_conditions
mean_weather_condition = round(df['weather_conditions'].mean())
df.loc[df['weather_conditions'] == -1, 'weather_conditions'] = mean_weather_condition
print("Mean weather condition after replacement:", mean_weather_condition)
```

###  Q2(a). Number of Motorcycles Accidents By Hour of The Day

```python
# Categorize motorcycles based on engine capacity
def categorize_motorcycle(cc):
    if cc == 2:
        return 'Motorcycle 50cc and under'
    elif cc == 3:
        return 'Motorcycle over 50cc and up to 125cc'
    elif cc == 4:
        return 'Motorcycle over 125cc and up to 500cc'
    elif cc == 5:
        return 'Motorcycle over 500cc'
    else:
        return 'Others'

df_copy['motorcycle_category'] = df_copy['vehicle_type'].apply(categorize_motorcycle)

# Filter accidents involving motorcycles
motorcycle_accidents = df_copy[df_copy['vehicle_type'].isin([2, 3, 4, 5])].copy()

# Convert 'time' column to datetime format
motorcycle_accidents['time'] = pd.to_datetime(motorcycle_accidents['time'])

# Extract hour of the day from the 'time' column
motorcycle_accidents['hour_of_day'] = motorcycle_accidents['time'].dt.hour

# Plot the distribution of accidents by hour of the day for each motorcycle category
plt.figure(figsize=(10, 6))
for category in df_copy['motorcycle_category'].unique():
    category_df = motorcycle_accidents[motorcycle_accidents['motorcycle_category'] == category]
    hourly_accidents = category_df['hour_of_day'].value_counts().sort_index()
    plt.plot(hourly_accidents.index, hourly_accidents.values, label=category)

plt.title('Distribution of Motorcycle Accidents by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.legend()
plt.grid(True)
plt.xticks(range(24))
plt.show()
```

### Q5. Identifying Regions and Clustering

```python
# Define the list of regions to filter
regions = ['Kingston upon Hull', 'Humberside', 'East Riding of Yorkshire']

# Filter accidents in the specified regions based on the 'lsoa01nm' column
region_df = df_copy[df_copy['lsoa01nm'].str.contains('|'.join(regions), case=False)]


region_df = region_df[region_df['accident_year'] == 2020]

# Display the filtered DataFrame
print(region_df)
```

```python
# Get the count of accidents after filtering
accident_count = region_df.shape[0]
print("Count of Accidents in the Specified Region(s):", accident_count)
```

```python
X = region_df[['latitude', 'longitude']]

# Determine the optimal number of clusters using the Elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the Elbow method to find the optimal number of clusters
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()

print('Optimal value of k is 3')
```

```python
# Based on the Elbow method, optimal number of clusters and performing K-means clustering
k_optimal = 3  
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
# Visualize the clusters on a scatter plot
plt.scatter(X['longitude'], X['latitude'], c=labels, cmap='viridis')
plt.scatter(centroids[:, 1], centroids[:, 0], marker='X', s=100, color='red', label='Centroids')
plt.title('Clustering of Accidents in the Specified Region')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Select relevant features for clustering
X = df_copy[['speed_limit', 'weather_conditions']]

# Perform K-means clustering
k_optimal = 15  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
labels = kmeans.fit_predict(X)

# Visualize the clusters and highlight different weather conditions
plt.figure(figsize=(10, 6))

# Plot each cluster with a different color based on weather condition
for i in range(k_optimal):
    cluster_indices = np.where(labels == i)[0]
    weather_conditions = X.iloc[cluster_indices]['weather_conditions']
    unique_conditions = weather_conditions.unique()
    for condition in unique_conditions:
        indices = cluster_indices[np.where(weather_conditions == condition)[0]]
        plt.scatter(X.iloc[indices]['speed_limit'], X.iloc[indices]['weather_conditions'], label=f'Cluster {i}, Weather {condition}')

plt.xlabel('Speed Limit')
plt.ylabel('Weather Conditions')
plt.title('Clusters of Accidents based on Speed Limits and Weather Conditions')
plt.show()
```

### Q(6). Outliers: IQR Method

```python
# Calculate IQR for longitude and latitude features
Q1_longitude = df_copy['longitude'].quantile(0.25)
Q3_longitude = df_copy['longitude'].quantile(0.75)
IQR_longitude = Q3_longitude - Q1_longitude

Q1_latitude = df_copy['latitude'].quantile(0.25)
Q3_latitude = df_copy['latitude'].quantile(0.75)
IQR_latitude = Q3_latitude - Q1_latitude

# Define thresholds for outliers
lower_bound_longitude = Q1_longitude - 1.5 * IQR_longitude
upper_bound_longitude = Q3_longitude + 1.5 * IQR_longitude

lower_bound_latitude = Q1_latitude - 1.5 * IQR_latitude
upper_bound_latitude = Q3_latitude + 1.5 * IQR_latitude

# Identify outliers
outliers_longitude1 = df_copy[(df_copy['longitude'] < lower_bound_longitude) | (df_copy['longitude'] > upper_bound_longitude)]
outliers_latitude1 = df_copy[(df_copy['latitude'] < lower_bound_latitude) | (df_copy['latitude'] > upper_bound_latitude)]
```

 ### Isolation Forest Method

```python
X_outlier = df_copy[['longitude', 'latitude']]

# Initialize the Isolation Forest model
isolation_forest = IsolationForest(random_state=42)

# Fit the model to the data and predict outliers
isolation_forest.fit(X_outlier)
outlier_labels = isolation_forest.predict(X_outlier)

# Identify outliers (outlier labels equal to -1)
outliers2 = df_copy[outlier_labels == -1]
```

### Local Outlier Factor Method

```python
X_outlier = df_copy[['longitude', 'latitude']]

# Initialize the Local Outlier Factor model
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)

# Fit the model to the data and predict outlier scores
outlier_scores = lof.fit_predict(X_outlier)

# Identify outliers (outlier scores less than -1 indicate outliers)
outliers3 = df_copy[outlier_scores == -1]
```
